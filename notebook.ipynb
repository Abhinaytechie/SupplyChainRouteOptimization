```python
import pulp
```

```python
supply={"W1":20,"W2":15}
demand={"S1":10, "S2":10, "S3":15}
costs={
        ("W1", "S1"): 8, ("W1", "S2"): 6, ("W1", "S3"): 10,
    ("W2", "S1"): 9, ("W2", "S2"): 12, ("W2", "S3"): 13
}
```

```python
model= pulp.LpProblem("Transportation_Problem",pulp.LpMinimize)
x=pulp.LpVariable.dicts("route",costs,lowBound=0,cat="Integer")
```

```python
model+=pulp.lpSum([ x[i,j] *costs[i,j] for (i,j) in costs])
```

```python
for i in supply:
    model+=pulp.lpSum([ x[i,j] for j in demand])<=supply[i]
```

```python
for i in demand:
    model+=pulp.lpSum([ x[j,i] for j in supply])== demand[i]
```

```python
model.solve()
```

```python
print("Status:", pulp.LpStatus[model.status])
for v in model.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
print("Total Cost =", pulp.value(model.objective))
```

```python
model2=pulp.LpProblem("Bakery_Problem",pulp.LpMaximize)
x=pulp.LpVariable("Bread",lowBound=0,cat="Continuous")
y=pulp.LpVariable("Cake",lowBound=0,cat="Continuous")
```

```python
from ortools.linear_solver import pywraplp
```

```python
solver =pywraplp.Solver.CreateSolver('CBC')
x=solver.NumVar(0,solver.infinity(), 'Bread')
y=solver.NumVar(0,solver.infinity(), 'Cake')
```

```python
solver.Add(2*x+4*y <=100)
solver.Add(x + 3*y <=80)
```

```python
solver.Maximize(2*x +5*y)
status=solver.Solve()
```

```python
if status== pywraplp.Solver.OPTIMAL:
    print("Optimal solution found:")
    print("Bread =", x.solution_value())
    print("Cake =", y.solution_value())
    print("Maximum Profit =", solver.Objective().Value())
```
